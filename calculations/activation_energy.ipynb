```python
import matplotlib as mlp
from matplotlib import pyplot as plt

import numpy as np
```

```python
R = 8.31 #J/K/mol
E_a = 80000 #J/mol = 88 kcal/mol
T = np. linspace(297, 630, 1000)
term = np.exp(-E_a / (R * T))
```

```python
fig, ax = plt.subplots(1,1)
ax.plot(T, term)
ax.set_yscale('log')
log_range = np.log10(term[0]) - np.log10(term[-1])
print(log_range)
```
